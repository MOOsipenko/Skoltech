```python
class Cell:
    
    def __init__(self, x_coordinate, y_coordinate):
        self.x = x_coordinate
        self.y = y_coordinate
        #self.state = state
        self.visited = 0
        self.cost = 0
        self.queue = []
        

def set_neighbours(x, y, visited):
    neighbours = []
    if x > 0:
        if [x-1, y] not in visited:
            neighbours.append([x-1, y])
            visited.append([x-1, y])
    if x < 99:
        if [x+1, y] not in visited:
            neighbours.append([x+1, y])
            visited.append([x+1, y])
    if y > 0:
        if [x, y-1] not in visited:
            neighbours.append([x, y-1])
            visited.append([x, y-1])
    if y < 99:
        if [x, y+1] not in visited:
            neighbours.append([x, y+1])
            visited.append([x, y+1])

    return neighbours
```

```python
def A_star_algorithm(X_I, X_G):
    
    Q = [X_I]
    X_I.visited = 1
    
    visited_cells = []
    
    while Q:
        
        # getting the item with the minimum cost (Priority Queue realisation)
        q = [obj for obj in Q if obj.cost == min(Q, key = lambda x: x.cost).cost][0]
        
        print('q:', q.x, q.y)
        
        if q.x == X_G.x and q.y == X_G.y:
            return list(q.cost, q.queue)
        
        current_neighbours = set_neighbours(q.x, q.y, visited_cells)
        current_cost = q.cost
        
        print("neighbours:", current_neighbours)
        
        for i in current_neighbours:
            c = Cell(i[0], i[1])
            
            if c.visited == 0:
                c.visited = 1
                c.cost = current_cost + 1
                c.queue.append(q)
                Q.append(c)
            else:
                if current_cost + 1 < c.cost:
                    c.cost = current_cost + 1
                    c.queue.append(q)
                    #Q.append(c)
            #print('Q:', '\n', Q)
            #print('c:', '\n', c.x, c.y)

        del Q[0]
                
        
    return -1
```
